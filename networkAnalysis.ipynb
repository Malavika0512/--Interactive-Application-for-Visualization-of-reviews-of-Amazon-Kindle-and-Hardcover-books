```python
Review_Per_Book = 20
Num_Of_Nodes=25
Num_Of_Books_Per_Genre=10
Review_Text_Column_Num =8
Asin_Column_Num=4

df1 = pd.DataFrame()
dataset = dataset1.sort_values(by ='Rank')
genreArray = dataset['Category'].unique()
finalNodesList ={}
finalEdgesList = {}
count = 0
for i in range(len(genreArray)):
    temp = dataset.loc[dataset['Category'] == genreArray[i]]
    #rank = temp['Rank'].unique()
    #temp = temp.loc[temp['Rank'].isin(rank[:Num_Of_Books_Per_Genre].tolist())]
    books = temp['Title'].unique()
    for j in range(len(books)):
        from nltk.corpus import stopwords
        stopwords = set(stopwords.words('english'))
        stopwords.add('also')
        
#         print len(stopwords)
#         print("Doing for genre " + str(i) + " book " + str(j))
        temp2 = temp.loc[temp['Title'] == books[j]]
        reviewText = pd.DataFrame(temp2).to_numpy()
        t = reviewText[:len(temp2),Review_Text_Column_Num]
#         for txt in text:
#             blob = TextBlob(txt)
#             #print(blob.noun_phrases)
#             nouns = list(blob.noun_phrases)
#             for n in nouns:
#                 stopwords.add(n)
#         print len(stopwords)
        tags = ['NN', 'PRP', 'PRP$', 'VB', 'VBD', 'WP', 'MD', 'RB', 'RBR', 'RBS']
        for txt in t:
            is_noun = lambda pos: pos[:2] in tags
            tokenized = nltk.word_tokenize(txt)
            nouns = [word for (word,pos) in nltk.pos_tag(tokenized) if is_noun(pos)]
            example_words = nouns
            filtered_sentence = []
            for w in example_words:
                if w not in stopwords:
                    x = unicode(w)
                    stopwords.add(x)        
   
        cv = CountVectorizer(ngram_range=(1,1), stop_words = stopwords)
        if len(temp2)>Review_Per_Book:
            X = cv.fit_transform(reviewText[:Review_Per_Book,Review_Text_Column_Num])
        else:
            X = cv.fit_transform(reviewText[:len(temp2),Review_Text_Column_Num])
        Xc = (X.T * X)
        Xc.setdiag(0)
        names = cv.get_feature_names()
        df = pd.DataFrame(data = Xc.toarray(), columns = names, index = names)
        for x in range(len(df.columns)-1,-1,-1):
            if df.columns[x].isdigit():
                df.drop(df.index[x],inplace=True)
                df.drop(df.columns[[x]],axis=1,inplace=True)
        print("Dropped integer column")
        result=[]
        for x in range(len(df)):
            temp3 = [x, sum(df.iloc[x])]
            result.append(temp3)
        resultArray = np.array(result)
        a= resultArray[resultArray[:,1].argsort()[::-1]]
        
        for x in range(len(df)-1,-1,-1):
            if x not in a[:Num_Of_Nodes,0]:
                df.drop(df.index[x],inplace=True)
                df.drop(df.columns[[x]],axis=1,inplace=True)
        print("Creating nodes and edges json")
        nodes = []
        for x in range(len(df)):
            nodes.append({"id" : df.index[x], "value" : sum(df.iloc[x])})
        finalNodesList.update({temp2.iloc[0,Asin_Column_Num] : nodes})
        
        edges = []
        for x in range(len(df)):
            for y in range(len(df.iloc[0])):
                edges.append({"source" : df.index[x], "target" : df.columns[y], "value": str(df.iloc[x][y])})
        finalEdgesList.update({temp2.iloc[0,Asin_Column_Num] : edges})
        break
    break

    #df1 = df1.append(temp)
#df1.to_csv("top10Books.csv",sep = ',',index=True)
###################################
# with open('nodes.json', 'w') as fp:
#     json.dump(finalNodesList, fp)
# fp.close
# with open('edges.json', 'w') as fp1:
#     json.dump(finalEdgesList, fp1)
# fp1.close
#########################################
    #df1 = df1.append(temp)
#df1.to_csv("top10Books.csv",sep = ',',index=True)
# with open('nodes.json', 'w') as fp:
#     json.dump(finalNodesList, fp)
# fp.close
# with open('edges.json', 'w') as fp1:
#     json.dump(finalEdgesList, fp1)
# fp1.close
```
