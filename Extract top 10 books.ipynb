```python
def returnUniqueIDs(dataset):
    year = dataset['Review_time'].unique()
    genre = dataset['Category'].unique()
    bookid = []
    for y in year:
        temp = dataset[dataset['Review_time'] == y]
        for g in genre:
            data = temp[temp['Category'] == g]
            data = data.sort_values(by = 'Rank')
            data = data.drop_duplicates(subset = ["Asin"])
            asins = list(data[:10]['Asin'])
            #print("year"+str(y)+"    genre:"+g + "       total asins:"+str(len(asins)))
            for a in asins:
                if a not in bookid:
                    bookid.append(a)
    print(len(bookid))
    return bookid
    
    
```

```python
def returnTop10(dataset):
    bookasins = returnUniqueIDs(dataset)
    top10 = pd.DataFrame()
    for i in bookasins:
        top10 = top10.append(dataset[dataset['Asin'] == i])
    return top10
        
```
